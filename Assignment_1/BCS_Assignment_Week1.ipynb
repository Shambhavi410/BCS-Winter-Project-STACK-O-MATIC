# **Part 3: Problem Solving with Python Basics**

### **Loops and Conditional Statements**

```python
#Fibonacci sequence up to 100 using a loop
e=0
f=1
L=[0,1]
while e<=100:
    g=f
    f=e+f
    if f<=100:
        L.append(f)
    else:
        break
    e=g
print("Fibonacci sequence up to 100:\n",L)
```

```python
#Loop through the numbers from 1 to 50 and print: "Fizz" for multiples of 3; "Buzz" for multiples of 5; "FizzBuzz" for multiples of both 3 and 5
for i in range(1,51):
    if (i%3==0 and i%5!=0):
        print("Fizz")
    elif (i%5==0 and i%3!=0):
        print("Buzz")
    elif (i%3==0 and i%5==0):
        print("FizzBuzz")
    else:
        print(i)
```

### **Variable-Length Arguments**

```python
#Write a solution that accepts a list of numbers and returns the median value
x=input("Enter a list of numbers separated by comma: ")
P=[]
for i in x.split(','):
      P.append(int(i))
P.sort()
y=len(P)
if y%2!=0:
    print("Median= ",P[((y+1)//2)-1])
else:
    print("Median= ",(P[(y//2)-1]+P[y//2])/2)
```

```python
#Write a solution that accepts a variable number of arguments and returns the average of the numbers
x=input("Enter a list of numbers separated by comma: ")
A=[]
for i in x.split(','):
        A.append(int(i))
print("Average= ",sum(A)/len(A))
```

### **Error Handling**

```python
#Handle the case where a division by zero occurs and print an appropriate error message.
num=int(input("Numerator: "))
den=int(input("Denominator: "))
if den==0:
    print("Division by 0 is not defined")
else:
    print(num/den)

#OR

try:
    result=num/den
    print(result)
except ZeroDivisionError:
    print("Division by 0 is not defined")
```

### **Advanced NumPy Manipulation**

```python
#Create a 3D NumPy array with specific dimensions and populate it with random numbers
s=np.random.rand(3,4,5)
print(s)
```

```python
#Swap two of its axes
#Let the three axes be (0,1,2)
#swapping axes 1 and 2
#The new 3D array with axes 1 and 2 swapped will be the one with the above three matrices transposed
new=np.zeros((3,5,4))
for i in range(3):
  for j in range(5):
    for k in range(4):
      new[i][j][k]=s[i][k][j]
print("Swapped axes 1 and 2 new 3D array:\n",new)
```

```python
#Swapping axes 0 and 2
new=np.zeros((5,4,3))
for i in range(5):
  for j in range(4):
    for k in range(3):
      new[i][j][k]=s[k][j][i]
print("Swapped axes 0 and 2 new 3D array:\n",new)
```

```python
#Swapping axes 0 and 1
new=np.zeros((4,3,5))
for i in range(4):
  for j in range(3):
    for k in range(5):
      new[i][j][k]=s[j][i][k]
print("Swapped axes 0 and 1 new 3D array:\n",new)
```

```python
#Perform an operation to calculate the sum of all the elements along a particular axis
#Let towards x axis there be 5 elements, towards y axis there be 4 elements and towards z axis there be 3 elements

#Sum along axis 0:
print("Sum along axis 0:\n",np.sum(s,axis=0))
#Sum along axis 1:
print("Sum along axis 1:\n",np.sum(s,axis=1))
#Sum along axis 2:
print("Sum along axis 2:\n",np.sum(s,axis=2))
```
